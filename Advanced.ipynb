```python
from transformers import pipeline
import torch
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
generator = pipeline('text-generation', model='gpt2')
```

```python
result = generator("Artificial Intelligence is", max_length=50, num_return_sequences=1)
print(result[0]['generated_text'])
```

```python
sample_texts = [
    "The future of artificial intelligence is",
    "Once upon a time in a distant galaxy,",
    "In the year 2050, humans discovered",
    "The most important lesson in life is"
]

for text in sample_texts:
    result = generator(text, max_length=50, num_return_sequences=1)
    print(f"Input: {text}")
    print(f"Generated Text: {result[0]['generated_text']}\n")
```

```python
result = generator(
    "Technology in 2050 will be",
    max_length=50,
    num_return_sequences=1,
    temperature=0.7,  
    top_k=50   
)

print(result[0]['generated_text'])
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import seaborn as sns

analyzer = SentimentIntensityAnalyzer()

sample_texts = [
    "AI is changing the world in positive ways.",
    "The consequences of artificial intelligence could be dangerous.",
    "I am unsure about the future of AI.",
    "The rapid advancements in AI will lead to a better society."
]

sentiments = [analyzer.polarity_scores(text)["compound"] for text in sample_texts]

sns.barplot(x=sample_texts, y=sentiments)
plt.xticks(rotation=45, ha="right")
plt.xlabel("Generated Text")
plt.ylabel("Sentiment Score (-1 to +1)")
plt.title("Sentiment Analysis of GPT-2 Generated Text")
plt.show()
```

```python
from wordcloud import WordCloud

generated_text = " ".join([
    generator("AI will", max_length=50)[0]['generated_text'],
    generator("Technology in 2050 will be", max_length=50)[0]['generated_text']
])

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(generated_text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Common Words in GPT-2 Generated Text")
plt.show()
```
